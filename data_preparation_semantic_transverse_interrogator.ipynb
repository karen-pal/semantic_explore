```python
import pandas as pd
import numpy as np
import tensorflow as tf
import tensorflow_hub as hub
from sklearn.manifold import TSNE
import plotly.graph_objs as go

# Load Universal Sentence Encoder model
use_model = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")

# Function to get USE embeddings
def get_use_embeddings(sentences):
    """
    Get semantic embeddings using Universal Sentence Encoder.
    """
    # Convert sentences to tensors
    sentence_tensors = tf.constant(sentences.tolist())

    # Get sentence embeddings
    sentence_embeddings = use_model(sentence_tensors).numpy()
    
    return sentence_embeddings

# Read DataFrame from CSV
df = pd.read_csv("full_dataset.csv")

# Apply embeddings function to all sentences
embeddings = get_use_embeddings(df['prompt'])

# Apply t-SNE to reduce dimensionality to 3 dimensions
tsne = TSNE(n_components=3, random_state=42)
embeddings_3d = tsne.fit_transform(embeddings)
```

```python
# Function to decode embeddings into sentences using the Universal Sentence Encoder
def decode_embedding_to_sentence(embedding):
    """
    Decode an embedding into a sentence using the Universal Sentence Encoder.
    """
    # Convert the embedding to a string representation
    embedding_str = ','.join(str(e) for e in embedding)

    # Convert the string to a tensor
    embedding_tensor = tf.constant([embedding_str])

    # Pass the embedding through the USE model to obtain the encoded sentence
    encoded_sentence = use_model(embedding_tensor)

    # Convert the tensor to a numpy array
    encoded_sentence_numpy = encoded_sentence.numpy()

    # Extract the encoded sentence string
    encoded_sentence_str = encoded_sentence_numpy[0]

    return encoded_sentence_str


def interpolate_3d_points(point1, point2, num_steps):
    """
    Interpolate linearly between two 3D points.
    """
    interpolated_points = []
    for i in range(num_steps):
        alpha = i / (num_steps - 1)
        interpolated_point = alpha * point1 + (1 - alpha) * point2
        interpolated_points.append(interpolated_point)
    return np.array(interpolated_points)

# Identify two points for interpolation
point_index1 = 310  # Index of the first point
point_index2 = 309  # Index of the second point

# Get the 3D embeddings of the identified points
embedding1 = embeddings_3d[point_index1]
embedding2 = embeddings_3d[point_index2]

# Interpolate between the embeddings
num_steps = 10  # Adjust as needed
interpolated_embeddings = interpolate_3d_points(embedding1, embedding2, num_steps)

# Decode interpolated embeddings into sentences
interpolated_sentences = []
for emb in interpolated_embeddings:
    # Decode each interpolated embedding into a sentence using the Universal Sentence Encoder
    # (You may need to adjust this part based on how you decode embeddings into sentences)
    interpolated_sentence = decode_embedding_to_sentence(emb)
    interpolated_sentences.append(interpolated_sentence)

# Print or display the interpolated sentences
for sentence in interpolated_sentences:
    print(sentence)
```

```python
import re

def get_plotly_div_id(html_file):
    with open(html_file, 'r') as f:
        html_content = f.read()
    match = re.search(r'<div id="(.*?)"', html_content)
    if match:
        return match.group(1)
    else:
        return None

# Example usage
html_file = "interactive_plot_with_path_color_and_toggleable_legend_and_log.html"
plotly_div_id = get_plotly_div_id(html_file)
print("Plotly div ID:", plotly_div_id)
```
