```python
import pandas as pd
import os

i = "12"
# Load the annotated dataset
annotated_dataset = pd.read_csv("annotated_full_dataset_USE+clustering.csv")

# Filter the dataset to get rows corresponding to cluster 0
cluster_data = annotated_dataset[annotated_dataset['cluster'] == int(i)].copy()
print("cluster:" , i)
# Remove ".csv" postfix from the path
cluster_data['path'] = cluster_data['path'].str.rstrip('.csv/')
cluster_data['path'] = cluster_data['path'].replace("acompañar","acompa__ar").replace("\n","")

# Get the file paths for the images in cluster 0
image_filepaths = "./videos/" + cluster_data['path'] + "/"+cluster_data['image']
#print(image_filepaths)
# Define output video file name
#output_video_file = "./cluster_0_video.mp4"

# Write the list of image file paths to a text file
with open("image_list_cluster"+i+".txt", 'w') as f:
    for image_filepath in image_filepaths:
        image_filepath = image_filepath.replace("ñ","__")
        f.write(f"file '{image_filepath}'\n")
        print(image_filepath)
# Use ffmpeg to concatenate images into a video
#os.system(f"ffmpeg -f concat -safe 0 -i image_list.txt -vf fps=8 -c:v libx264 -r 30 -pix_fmt yuv420p {output_video_file}")
```

```python
import cv2

# Read the list of image paths from the text file
with open('image_list_cluster'+i+'.txt', 'r') as file:
    image_paths = file.readlines()
print(i)
# Remove newline characters from the end of each path
image_paths = [path.replace("file ","").replace("'","").replace("\n","") for path in image_paths]
print(image_paths[0])
# Create a VideoWriter object to save the video
video_writer = cv2.VideoWriter('output_opencv_cluster'+i+'.mp4', cv2.VideoWriter_fourcc(*'mp4v'), 8, (640, 480))

# Iterate through each image path and add it to the video
for image_path in image_paths:
    # Read the image
    image = cv2.imread(image_path)
    if image is not None:
        # Resize the image to match the desired video dimensions
        image = cv2.resize(image, (640, 480))
        # Write the image to the video
        video_writer.write(image)

# Release the VideoWriter object
video_writer.release()
```
